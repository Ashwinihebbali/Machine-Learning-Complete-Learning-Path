```python
# Concept Learning Based Book Recommendation

history = {
    "S1": ["AI","Machine Learning","Deep Learning"],
    "S2": ["Python","Data Science","AI"],
    "S3": ["Fiction","Novel","Poetry"],
    "S4": ["Machine Learning","AI"],
    "S5": ["Data Science","Python","ML Basics"]
}

# Map titles to broader concepts (types)
types = {"AI":"Tech","Machine Learning":"Tech","Deep Learning":"Tech","ML Basics":"Tech",
         "Python":"Tech","Data Science":"Tech",
         "Fiction":"Literature","Novel":"Literature","Poetry":"Literature"}

# Learn the concept: who is a "Tech enthusiast"?
concept = {}
for books in history.values():
    cat = [types[b] for b in books]
    for c in cat:
        concept[c] = concept.get(c, 0) + 1

# Dominant concept = category borrowed most often
dominant = max(concept, key=concept.get)

# Recommend books matching the learned concept
def recommend(current_books):
    cur = {types[b] for b in current_books}
    if dominant in cur:
        return [b for b,t in types.items() if t == dominant and b not in current_books]
    else:
        return [b for b,t in types.items() if t != dominant and b not in current_books]

print("Learned concept:", dominant)
print("For ['AI'] ->", recommend(["AI"]))
print("For ['Fiction'] ->", recommend(["Fiction"]))
```
