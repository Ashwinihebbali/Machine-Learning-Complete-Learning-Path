```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#1. Load the dataset
df = pd.read_csv("food_preferences_survey.csv")

#2. Univariate Analysis
print("Frequency of Food Items:")
print(df['preferred_food'].value_counts())

#3. Bar Chart
sns.countplot(x='preferred_food', data=df)
plt.title("Popular Food Items")
plt.show()

#4. Pie Chart
df['preferred_food'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Food Preferences Distribution")
plt.ylabel("")
plt.show()

#5. Bivariate Analysis: Gender vs Preferred Food
print("\nGender vs Preferred Food:")
print(pd.crosstab(df['gender'], df['preferred_food']))

sns.countplot(x='preferred_food', hue='gender', data=df)
plt.title("Gender vs Preferred Food")
plt.show()

#6. Bivariate Analysis: Year vs Preferred Food
print("\nYear of Study vs Preferred Food:")
print(pd.crosstab(df['year_of_study'], df['preferred_food']))

sns.countplot(x='preferred_food', hue='year_of_study', data=df)
plt.title("Year of Study vs Preferred Food")
plt.show()
```
