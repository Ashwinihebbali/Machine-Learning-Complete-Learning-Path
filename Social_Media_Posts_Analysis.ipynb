```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# 1. Load the dataset
data = pd.read_csv("social_media_posts.csv")

# 2. Lowercase
data["cleaned_text"] = data["post_text"].str.lower()

# 3. Remove punctuation and stopwords
punct = [",", ".", "!"]
stop = ["the", "is", "and", "for", "to"]

cleaned = []
for txt in data["cleaned_text"]:
    for p in punct:
        txt = txt.replace(p, " ")
    words = txt.split()
    words = [w for w in words if w not in stop]
    cleaned.append(" ".join(words))

data["cleaned_text"] = cleaned

# 4. Top keywords
all_words = " ".join(data["cleaned_text"]).split()
word_counts = Counter(all_words)
print("Top 10 Keywords:\n", word_counts.most_common(10))

# 5. Event type counts
event_counts = data["event_type"].value_counts()
print("\nEvent Type Counts:\n", event_counts)

# 6. Bar chart
event_counts.plot(kind="bar", color="skyblue", title="Most Discussed Event Types")
plt.xlabel("Event Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
